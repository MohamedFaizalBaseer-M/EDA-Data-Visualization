```python
#Label Encoding for Category and Content Rating
#OneHotEncoding for Type and Rating_category
le_col=['Category','Content Rating','Rating_category']
ohe_col=['Type']
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
for i in le_col:
    ds[i]=le.fit_transform(ds[i])
```

```python
#OneHotEncoding
ohe=OneHotEncoder(sparse=False)
ohe_ds=pd.DataFrame(ohe.fit_transform(ds[ohe_col]))
ohe_ds
```

```python
a=ohe.categories_
ohe_ds.columns=list(a[0])
ohe_ds
```

```python
ds.reset_index(drop=True,inplace=True)
```

```python
ds_final=pd.concat((ds,ohe_ds),axis='columns')
```
